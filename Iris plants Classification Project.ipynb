split data into training set and test set 

with 70% training set (105) and 30% test set (45)

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.30, random_state=42)
```

check the shape of the training set and test set

```python
print(X_train.shape)
print(y_train.shape)

print(X_test.shape)
print(y_test.shape)
```

Let's import several models (Logistic_Regression, SVC, Decision_Tree_Classifier, Gradient_Boosting_Classifier) 
and store our models into dictionary

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier

models = {
    'Logistic_Regression' : LogisticRegression(random_state=42),
    'SVC' : SVC(kernel='rbf', gamma=0.5, probability=True),
    'Decision_Tree_Classifier' : DecisionTreeClassifier(max_depth=2, random_state=42),
    'Gradient_Boosting_Classifier' : GradientBoostingClassifier(max_depth=1, n_estimators=3, learning_rate=1.0, random_state=42)
}
```

import accuracy_score metric for mesuring how good our models perform on the dataset

and creat the function that shows the accuracy score

```python
from sklearn.metrics import accuracy_score

def print_score(y_true,y_pred):
    print("accuracy_score :", accuracy_score(y_true,y_pred))
    print('*'*50)
```

creat the function that trains our models, shows models' names and calls the print_score function to see our models' metrics 

```python
def train_models(models,X,y):
  for name,model in models.items():
    print(name," : ")
    model.fit(X,y)
    y_pred = model.predict(X)
    print_score(y,y_pred)

train_models(models,X_train,y_train);
```

As we can see, SVC can perform well on the training dataset than other models

with accuracy_score = 0.9904761904761905

However, this is the score on the training dataset. we have to test our model on the test dataset. Let's go for it.

```python
def test_models(models,X,y):
  for name,model in models.items():
    print(name," : ")
    y_pred = model.predict(X)
    print_score(y,y_pred)
```

```python
test_models(models,X_test,y_test)
```

Logistic_Regression, SVC and Gradient_Boosting_Classifier can perform well on this dataset with their accuracy scores = 1.0 amazing!!

we pick the Logistic_Regression as our selected model due to its performance (accuracy = 1.0).

we compare the predicted values(prediction of X_test) and actual values(y_test) of Logistic_Regression model.

```python
models['Logistic_Regression'].predict(X_test)
```

```python
y_test
```

As we can see above, predicted values(prediction of X_test) match actual values(y_test) perfectly.this verifies the accuracy score (1.0) of the model.

next, we save the model using joblib

```python
import joblib
joblib.dump(models['Logistic_Regression'], 'model.pkl')
```

next. we load the saved model and test it

```python
logistic_regression_model = joblib.load("model.pkl")

logistic_regression_model.predict(X_test)
```
